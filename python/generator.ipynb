https://realpython.com/introduction-to-python-generators/ 읽고 실습을 진행하는 파일입니다.


아래 코드는 무식하게 csv 파일의 모든 내용을 메모리에 적재한 뒤에 행 개수를 일일이 세는 프로그램입니다.


```python
import csv
% % timeit


def csv_reader(file_name):
    file = open(file_name)
    result = file.read().split("\n")
    return result


csv_gen = csv_reader("resources/techcrunch.csv")
row_count = 0

for row in csv_gen:
    row_count += 1

# print(f"row count = {row_count}")
```

아래 코드는 Lazy하게 그때그때 파일에서 행을 읽어들이는 방식을 사용하여 메모리 초과 우려가 적습니다. 심지어 필요한 시간도 줄어든다...!

파일을 곧바로 이터레이트하면 행 단위로 끊는다는 사실!


```python
% % timeit


def csv_reader_generator(file_name):
    for row in open(file_name, 'r'):
        yield row


csv_gen = csv_reader_generator("resources/techcrunch.csv")
row_count = 0

for row in csv_gen:
    row_count += 1

# print(f"row count = {row_count}")
```

generator comprehension이라는 것도 있네! list comprehension과 모습은 완전 동일하다.


```python
% % timeit

csv_gen = (row for row in open("resources/techcrunch.csv"))
row_count = 0

for row in csv_gen:
    row_count += 1
```

## Palindrome detector with generator function

팰린드롬은 '우영우'같이 앞뒤가 똑같은 것들을 말한다. 이번엔 팰린드롬 숫자를 검출하는 코드를 작성해보자.


```python
def is_palindrome(num):
    # skip 0-ended numbers
    if num // 10 == 0:
        return False

    tmp = num
    reversed_sum = 0

    while tmp > 0:
        reversed_sum = (reversed_sum * 10) + (tmp % 10)
        tmp //= 10

    if num == reversed_sum:
        return True
    else:
        return False


def infinite_sequence():
    i = 0
    while True:
        yield i
        i += 1


for i in infinite_sequence():
    if i > (1 << 21):
        break
    if is_palindrome(i):
        print(i)
```
